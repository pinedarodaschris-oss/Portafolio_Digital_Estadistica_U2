### Tarea APE 09: Pruebas de Hipótesis Unimuestrales

**Descripción:** Defina una hipótesis nula ($H_0$) y una alterna ($H_1$) sobre un parámetro crítico de la región de Loja. Ejecute un test paramétrico (Z o T de Student) y justifique estadísticamente la decisión basada en el Valor-p obtenido.

Para realizar una prueba de hipótesis unimuestral, primero necesitamos identificar una variable numérica relevante en el dataset y un valor de referencia (media o proporción) para la hipótesis.

```python
from scipy import stats
import numpy as np

# Filtrar datos para la provincia de Loja
df_loja = df_dengue[df_dengue['Provincia'] == 'LOJA']

# Seleccionar la variable 'Total' para el análisis
total_cases_loja = df_loja['Total']

total_cases_loja = total_cases_loja.dropna()

# Calcular la media muestral de los casos totales en Loja
mean_total_loja = total_cases_loja.mean()
std_total_loja = total_cases_loja.std()
n_total_loja = len(total_cases_loja)

print(f"Número de registros para Loja: {n_total_loja}")
print(f"Media de casos totales en Loja: {mean_total_loja:.2f}")
print(f"Desviación estándar de casos totales en Loja: {std_total_loja:.2f}")

# Definir la hipótesis nula y alterna
# H0: La media de casos totales de dengue en Loja es igual a un valor hipotético (ej. 2 casos por registro).
# H1: La media de casos totales de dengue en Loja es diferente de 2 casos por registro.

# Establecemos un valor hipotético (mu) para la prueba. Podemos ajustarlo segùn un contexto o interés específico.
# Por ejemplo, podemos hipotetizar que la media de casos por registro es 2.
mu_hipotetica = 2

# Realizar la prueba t de una muestra
t_statistic, p_value = stats.ttest_1samp(total_cases_loja, mu_hipotetica)

print(f"\nHipótesis Nula (H0): La media de casos totales de dengue en Loja es igual a {mu_hipotetica}.")
print(f"Hipótesis Alterna (H1): La media de casos totales de dengue en Loja es diferente de {mu_hipotetica}.")
print(f"Estadístico t: {t_statistic:.4f}")
print(f"Valor p: {p_value:.4f}")

# Nivel de significancia
alpha = 0.05

print(f"Nivel de significancia (alpha): {alpha}")
if p_value < alpha:
    print("Conclusión: Se rechaza la hipótesis nula. Hay evidencia suficiente para decir que la media de casos totales de dengue en Loja es diferente de {mu_hipotetica}.")
else:
    print("Conclusión: No se rechaza la hipótesis nula. No hay evidencia suficiente para decir que la media de casos totales de dengue en Loja es diferente de {mu_hipotetica}.")
```

### Justificación Estadística de la Decisión

El **Valor-p** es la probabilidad de observar un estadístico de prueba (en este caso, el estadístico t) tan extremo o más extremo que el obtenido de la muestra, asumiendo que la hipótesis nula ($H_0$) es verdadera. En términos más simples, nos ayuda a determinar si nuestros datos son inusuales bajo la suposición de que $H_0$ es correcta.

*   Si el **Valor-p < $\alpha$** (nivel de significancia, comúnmente 0.05), entonces consideramos que los resultados son estadísticamente significativos. Esto significa que es poco probable que hayamos observado nuestros datos si $H_0$ fuera verdadera, por lo que **rechazamos $H_0$**. Concluimos que hay evidencia a favor de la hipótesis alternativa ($H_1$).

*   Si el **Valor-p $\ge$ $\alpha$**, entonces los resultados no son estadísticamente significativos. No tenemos suficiente evidencia para rechazar $H_0$. Esto **no significa que $H_0$ sea verdadera**, sino simplemente que nuestros datos no proporcionan pruebas suficientes para afirmar lo contrario.

### Comparación de Grupos (Tareas APE 10 y 11: A/B Testing o ANOVA)

**Descripción:** Divida su dataset regional en subgrupos de análisis (ej. datos de diferentes microclimas o sectores de Loja). Implemente un algoritmo de A/B Testing o un Análisis de Varianza (ANOVA de 1 factor) con su respectiva prueba Post-Hoc de Tukey para determinar si existen diferencias significativas entre las medias de los grupos.

Para esta sección, realizaremos un Análisis de Varianza (ANOVA) de un factor para comparar las medias de casos totales de dengue entre diferentes cantones dentro de la provincia de Loja. Primero, necesitamos identificar al menos tres cantones para el análisis y luego extraer sus datos.

```python
# Identificar cantones óptimos para el análisis ANOVA en Loja
cantones_en_loja = df_loja['Canton'].value_counts()
print("Cantones en la provincia de Loja y su recuento de registros:\n", cantones_en_loja)

# Seleccionar al menos tres cantones con suficientes datos para el análisis
# Elegiremos los cantones con mayor número de registros para asegurar robustez.
cantones_seleccionados = cantones_en_loja[cantones_en_loja >= 30].index.tolist()

# Si hay más de 3 cantones que cumplen el criterio, seleccionar los 3 o 4 primeros.
if len(cantones_seleccionados) > 4:
    cantones_para_anova = cantones_seleccionados[:4]
elif len(cantones_seleccionados) > 2:
    cantones_para_anova = cantones_seleccionados
else:
    # Si no hay suficientes cantones con al menos 30 registros, se eligen los 3 primeros de la lista total
    cantones_para_anova = cantones_en_loja.index.tolist()[:3]

print(f"\nCantones seleccionados para ANOVA: {cantones_para_anova}")

# Extraer los datos de 'Total' para cada cantón seleccionado
grupos_anova = []
for canton in cantones_para_anova:
    # Filtrar df_loja para el cantón actual y seleccionar la columna 'Total'
    casos_canton = df_loja[df_loja['Canton'] == canton]['Total'].dropna()
    grupos_anova.append(casos_canton)
    print(f"Registros de casos totales para {canton}: {len(casos_canton)}")

# Realizar la prueba ANOVA de un factor
# H0: Las medias de casos totales de dengue son iguales en todos los cantones seleccionados.
# H1: Al menos una media de casos totales de dengue es diferente entre los cantones seleccionados.

f_statistic, p_value_anova = stats.f_oneway(*grupos_anova)

print(f"\n\nRealizando ANOVA para los cantones: {', '.join(cantones_para_anova)}")
print(f"Hipótesis Nula (H0): $\mu_1 = \mu_2 = ... = \mu_k$ (Las medias de casos totales son iguales en todos los cantones seleccionados)")
print(f"Hipótesis Alterna (H1): Al menos una media de casos totales es diferente entre los cantones seleccionados")

print(f"Estadístico F: {f_statistic:.4f}")
print(f"Valor p (ANOVA): {p_value_anova:.4f}")

# Nivel de significancia para ANOVA
alpha_anova = 0.05

print(f"Nivel de significancia (alpha): {alpha_anova}")
if p_value_anova < alpha_anova:
    print("Conclusión: Se rechaza la hipótesis nula. Hay evidencia estadística para sugerir que al menos una de las medias de casos totales de dengue es significativamente diferente entre los cantones seleccionados.")
    print("Se procederá con la prueba Post-Hoc de Tukey para identificar qué pares de cantones difieren.")
else:
    print("Conclusión: No se rechaza la hipótesis nula. No hay evidencia estadística para sugerir diferencias significativas entre las medias de casos totales de dengue de los cantones seleccionados.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd

# Preparar los datos para la prueba Post-Hoc de Tukey
# Necesitamos un solo DataFrame con una columna para los valores y otra para los grupos.

data_for_tukey = pd.DataFrame()
all_cases = []
all_cantons = []

for i, canton_name in enumerate(cantones_para_anova):
    cases = grupos_anova[i] # grupos_anova contiene las Series de 'Total' para cada cantón
    all_cases.extend(cases)
    all_cantons.extend([canton_name] * len(cases))

data_for_tukey['Total_Cases'] = all_cases
data_for_tukey['Canton'] = all_cantons

print("\nRealizando la prueba Post-Hoc de Tukey HSD...")

tukey_result = pairwise_tukeyhsd(endog=data_for_tukey['Total_Cases'], # Variable dependiente
                                 groups=data_for_tukey['Canton'],    # Variable de agrupación
                                 alpha=alpha_anova)                  # Nivel de significancia

print(tukey_result)
```

### Prueba Post-Hoc de Tukey HSD: Interpretación

La prueba Post-Hoc de Tukey HSD (Honestly Significant Difference) se utiliza después de una ANOVA cuando la hipótesis nula ha sido rechazada. Su propósito es identificar cuáles pares específicos de medias de grupo son significativamente diferentes entre sí.

Los resultados de Tukey HSD se presentan generalmente en una tabla que muestra:

*   **group1 vs. group2**: Los pares de grupos que se están comparando.
*   **meandiff**: La diferencia entre las medias de los dos grupos.
*   **p-adj**: El valor p ajustado para mùltiples comparaciones. Este es el valor clave para la interpretación.
*   **lower, upper**: El intervalo de confianza del 95% para la diferencia de medias.
*   **reject**: Un valor booleano (True/False) que indica si la hipótesis nula de no diferencia entre las medias de ese par de grupos se rechaza o no.

**Criterio de Decisión:**

*   Si **p-adj < $\alpha$** (nivel de significancia, que aquí es 0.05), entonces **rechazamos la hipótesis nula** para ese par de grupos. Esto significa que hay una diferencia estadísticamente significativa entre las medias de esos dos grupos.
*   Si **p-adj $\ge$ $\alpha$**, entonces **no rechazamos la hipótesis nula** para ese par de grupos. No hay evidencia estadísticamente significativa para afirmar que las medias de esos dos grupos son diferentes.

### Interpretación del Valor-p en ANOVA

Al igual que en la prueba t, el **Valor-p** en ANOVA nos indica la probabilidad de obtener un estadístico F tan extremo o más extremo que el observado, si la hipótesis nula ($H_0$) de que todas las medias de los grupos son iguales fuera cierta.

*   **Si el Valor-p (ANOVA) < $\alpha$** (nivel de significancia), **rechazamos la hipótesis nula**. Esto significa que existe al menos un par de grupos cuyas medias son significativamente diferentes. Sin embargo, ANOVA no nos dice *cuáles* grupos son diferentes, solo que *al menos uno* lo es. Para identificar los grupos específicos, necesitamos una prueba Post-Hoc.

*   **Si el Valor-p (ANOVA) $\ge$ $\alpha$**, **no rechazamos la hipótesis nula**. Esto sugiere que no hay evidencia estadísticamente significativa para afirmar que las medias de los grupos son diferentes. En este caso, no sería necesario realizar una prueba Post-Hoc.

En nuestro caso, el Valor-p de la prueba ANOVA nos guiará en si procedemos con una prueba Post-Hoc de Tukey.
